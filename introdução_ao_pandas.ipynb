# Ordenando os dados

Para ordenar um DataFrame, use o método `sort_values()`. A maneira mais simples de uso é passando o nome de uma coluna entre parênteses.

````python
df.sort_values('SALARY')
````

Como padrão esse método ordena de forma crescente. Se desejar que seja decrescente, acrescente `ascending=False`. Observe que o valor boleano não é uma string, por isso não está entre aspas e tem a letra inicial em maiúscula. <br>Alguns métodos ou funções possuem diversos argumentos que podem ser passados entre parênteses, para saber quais são consulte a [documentação do pandas](https://pandas.pydata.org/docs/). Se desejar passar mais de um argumento, basta separar eles com vírgulas.

````python
df.sort_values('SALARY', ascending=False)
````

Você pode encadear vários métodos em uma linha. Por exemplo, se quisessemos ordenar os dados de maneira descendente e mostrar apenas as cinco primeiras linhas, poderíamos fazer dessa forma:

````python
df.sort_values('SALARY', ascending=False).head()
````

Você pode ordenar várias colunas de uma vez, passando uma lista com nome de colunas ao invés de uma única coluna. Uma lista é um conjunto de itens separado por vírgulas, entre colchetes `[]`.<br>
Para ordenar primeiro `SALARY` e depois `TEAM`: 

````python
df.sort_values(['SALARY','TEAM']).head()
````

Você pode especificar o ordenamento (ascendente ou descendente) para cada coluna listada, passando um outra lista para o `ascending` com `True` e `False`, na ordem das colunas a serem ordenadas. Por exemplo, para ordenar a coluna `SALARY` como descendente e `TEAM` como `ascendente`, seria:

````python
df.sort_values(['SALARY', 'TEAM'], ascending=[False,True]).head()
````

O `False` foi para coluna `SALARY` e o `True` para coluna `TEAM` porque ele estão na mesma posição que as colunas foram passadas na lista.<p>
**Observação:** Apesar de ter feito todas essas ordenações, nenhuma alteração foi aplicada ao dataframe original.


````python
df.head()
````

Isso acontece porque não salvamos os resultados do ordenamento em uma variável. Se você quiser manter o ordenamento que você fez (ou qualquer outro tipo de manipulação), recomenda-se guardar os resultados em uma variável.

````python
ordenar_por_team = df.sort_values('TEAM')
````

````python
ordenar_por_team.head()
````

# Filtrar os dados

Vamos vê dois tipos diferentes de filtro:


*   **Filtro de coluna:** Trás uma ou mais colunas de dados especificadas, semelhante a passar nome de colunas no `SELECT` em uma consulta SQL.
*   **Filtro de linha:** Trás um recorte dos seus dados que atenda aos critérios que foi especificado, semelhante a uma consulta SQL usando `WHERE`. Por exemplo: Me mostre todos os dados no meu DataFrame onde os valores da coluna `TEAM` seja igual a `ARI`. 



**Filtro de coluna**<p>
Para acessar os valores em uma única coluna de dados, você pode usar um ponto e o nome da coluna, desde que o nome da coluna não tenha espaços ou caracteres especiais.

````python
df.TEAM
````

Outro jeito é usando os colchetes com o nome da coluna entre aspas, equivalente ao comando a seguir:

````python
df['TEAM']
````

Quando você acessa uma única coluna no seu DataFrame, você está pegando uma parte dele chamado de `Series`. <p>
Um dos métodos que podem ser chamados em uma Series é `unique()`, que mostra apenas valores únicos da coluna. Vamos fazer isso com a coluna `TEAM`:

````python
df.TEAM.unique()
````

O que fizemos é a mesma coisa que pegar conteúdo da coluna `TEAM` para as linhas de uma tabela dinâmica, ou em SQL:

```sql

FROM mlb
```
Você pode contar o total de cada valor usando o método `values_counts()`:

````python
df.TEAM.value_counts()
````

*   `min()` pega o menor valor
*   `max()` pega o maior valor
*   `median()` pega a mediana
*   `mean()` pega a média
*   `mode()` pega o valor mais comum <p>
Vamos checar esses métodos, aplicando na coluna `SALARY`: 



````python
df.SALARY.min()
````

````python
df.SALARY.max()
````

````python
df.SALARY.median()
````

````python
df.SALARY.mean()
````

````python
df.SALARY.mode()
````

**Linhas filtradas**

O colchetes também é usado para filtrar linhas, exceto quando ao invés do nome da coluna (ou a lista de nomes) seja colocado dentro do colchetes uma condição. <p>

Vamos filtrar nossos dados para vê qual jogador tem salário maior que 1000000 dólares, ou seja, o filtro deve retornar as linhas onde o valor da coluna `SALARY` seja maior que $1.000.000 dólares.

````python
df[df.SALARY > 1000000]
````

Esse filtro em SQL seria:<br>


É possível usar os operadores de comparação do Python:


*   `>`  maior que

*   `<`  menor que
*   `<=` menor ou igual a

# Agupar e agregar os dados

O método `groupby` é usado para agrupar e agregar os dados, semelhante ao que é feito em uma tabela dinâmica ou uma consulta SQL. O Pandas também tem um método para tabela dinâmica, caso lhe interesse depois [você pode conferir na documentação](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html). <p>
Digamos que queremos ver o top 10 dos times por folha de pagamento. Em outras palavras nós queremos:


*   Agrupar os dados pela coluna `TEAM`: `groupby()`
*   Somar os resultados de cada grupo: `sum()`
*   Ordenas os resultados pela coluna `SALARY` de forma descendente: `sort_values()`
*   Mostrar apenas o top 10: `head(10)`<p>
Chamar o método `groupby()` sem dizer o que ele faz com os resultados agrupados não ajuda muito:

````python
df.groupby('TEAM')
````

Basicamente o Pandas só nos diz que o agrupamento foi bem sucedido. O que fazemos agora? Usando métodos encadeados, descrevemos o que gostariamos de fazer com as colunas numéricas depois que elas estão agrupadas. <p>Vamos começar com `sum()`:

````python
df.groupby('TEAM').sum()
````

Perfeito! Exceto pelo fato que o Pandas somou todas as colunas numéricas e não apenas a coluna `SALARY`. Para ajustar isso, filtraremos as duas colunas que estamos interessados - agrupando pela coluna `TEAM` e então somaremos os valores. Como nesse filtro a única coluna numérica é `SALARY`, o Pandas vai entender que é apenas essa que deve ser somada.

**Lembre-se:** Para selecionar as colunas de um dataframe, use colchetes e liste o nome das colunas, separando com vírgula.





````python
df[['TEAM', 'SALARY']].groupby('TEAM').sum()
````

Ótimo! Agora encadeando os métodos vamos ordenar a coluna `SALARY` de maneira descendente, trazendo como resultado apenas o top 10:

````python
df[['TEAM','SALARY']].groupby('TEAM').sum().sort_values('SALARY', ascending=False).head(10)
````

Você pode agregar outros métodos além do `sum()`, como `mean()` e `median()`, ou usar o método `agg()` para aplicar mais de uma operação.

````python
df[['TEAM','SALARY']].groupby('TEAM').median()
````

````python
df[['TEAM','SALARY']].groupby('TEAM').mean()
````

````python
df[['TEAM','SALARY']].groupby('TEAM').agg(['sum','mean','median'])
````
